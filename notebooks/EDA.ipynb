ðŸ‘† we can see overview column is in string format. let's convert this to list.\ will do using `lambda` and `split` function

```python
df['overview'] = df['overview'].apply(lambda x: x.split())
```

```python
df.head()
```

#### we will merge words of single name or genres
#### like science fiction to sciencefiction 

```python
df['genres'] = df['genres'].apply(lambda gen : [item.replace(" ", "") for item in gen])
```

### same thing for `crew`, `cast`, `keywords`

```python
df['cast'] = df['cast'].apply(lambda c : [item.replace(" ", "") for item in c])
df['crew'] = df['crew'].apply(lambda c : [item.replace(" ", "") for item in c])
df['keywords'] = df['keywords'].apply(lambda k : [item.replace(" ", "") for item in k])
```

```python
df.head()
```

## now we will merge all the columns except title and movie_id to a column named tags to create tags for recommender system

```python
df['tags'] = df['cast'] + df['keywords'] + df['crew'] + df['genres']
```
